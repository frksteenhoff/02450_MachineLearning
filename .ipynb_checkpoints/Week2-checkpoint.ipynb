### Principal Component Analysis

```python
from scipy.linalg import svd
```

```python
# Subtract mean value from data
Y = Xpandas - Xpandas.mean(0) # changed in order to fit the pandas format

# PCA by computing SVD of Y
U,S,V = svd(Y,full_matrices=False)

# Compute variance explained by principal components
rho = (S*S) / (S*S).sum() 

# Plot variance explained
figure()
plot(range(1,len(rho)+1),rho,'o-')
title('Variance explained by principal components');
xlabel('Principal component');
ylabel('Variance explained');
show()
```

```python
pd.DataFrame(rho)
```

**The variance explained by principal components**

Usually this is calculated from the diagonal matrix, but here *rho* already gives the variance explained. Using these values and the graph we can see that the 1st principal component accounts for approx. $76%$, 2nd $11.5%$ and the 3rd $4.8%$. This sums up to a total of $76+11.5+4.8 = 92.3%$ which is more than $90%$ as what was expected from the assignment.

### PCA 1 and 2 in scatterplot

```python
V = V.T
# Project the centered data onto principal component space
Z = np.dot(Y,V)

# Indices of the principal components to be plotted
i = 0
j = 1

# Plot PCA of the data
f = figure(figsize=(10, 7))
f.hold()
title('NanoNose data: PCA')
#Z = array(Z)

for c in range(C):
    # select indices belonging to class c:
    class_mask = y.A.ravel()==c
    plot(Z[class_mask,i], Z[class_mask,j], 'o')
legend(classNamesPandas)

xlabel('PC{0}'.format(i+1))
ylabel('PC{0}'.format(j+1))

# Output result to screen
show()
```

#### What are the benefits of visualizing the data by the projection given by PCA over plotting two of the original data dimensions against each other? Compare with the scatter plots of attributes you made in Exercise 2.1.2.

The projection of the data from the PCA aligns the directions of the datapoints such that the different principal components overlap as little as possible -- the optimal split has been chosen. Makes it easier to read.

```python
# Digits to include in analysis (to include all, n = range(10) )
n = [0,1]
# Number of principal components for reconstruction
K = 100
# Digits to visualize
nD = range(6);


# Load Matlab data file to python dict structure, extract variables of interest
traindata = loadmat('Data/zipdata.mat')['traindata']
X = traindata[:,1:]
y = traindata[:,0].T

N,M = X.shape
C = len(n)

classValues = np.unique(y)
classNames  = [str(int(v)) for v in classValues]
classDict   = dict(zip(classNames,classValues))
```

```python
# Select subset of digits classes to be inspected
class_mask = np.zeros(N).astype(bool) # bool array

for v in n:
    cmsk = (y == v)
    class_mask = class_mask | cmsk
X = X[class_mask,:]
y = y[class_mask]
N=X.shape[0]

# Center the data (subtract mean column values)
# Xc = Y
Xc = X - np.ones((N,1))*X.mean(0)
```

```python
# PCA by computing SVD of Y
# S = singular values rho^2
# V = unitary matrix
# U = 
U,S,V = np.linalg.svd(Xc,full_matrices=False)
#U = mat(U)
V = V.T

# Compute variance explained by principal components
rho = (S*S) / (S*S).sum() 

# Project data onto principal component space
Z = np.dot(Xc,V)
```

### Visualization

```python
# Plot variance explained
figure(figsize=(12,7))
plot(rho,'o-')
title('Variance explained by principal components');
xlabel('Principal component');
ylabel('Variance explained value');
```

**Show that it requires 22 PCA components to account for more than 90% of the variance in the data.**

```python
# Checking the actual value for the first 22 components
sum(rho[:22])
```

Actually 23 PCA components is needed in order to account for more than $90$% of the variance in the data -- python is 0 indexed and using the first 22 only accounts for $89.5$%.

```python
# Plot PCA of the data
f = figure(figsize=(10,7))
f.hold()
title('NanoNose data: PCA')
for c in n:
    # select indices belonging to class c:
    class_mask = (y == c)
    plot(Z[class_mask,0], Z[class_mask,1], 'o')
legend(classNames)
xlabel('PC1')
ylabel('PC2')
show()
```

**Show that the 1st principal component almost is sufficient for describing the data **

Looking at the visualization using both the first and the second principal component we can see that the zeros and ones are split into two (almost) separate clusters one to the left, the onesm and more towards the right, the zeros  -- with quite a larger variance. Only a few data points overlap. If we were to plot this in one dimension the result would be a single line with two clusters. A visualization of this has been made below.

```python
# plotting data points in one dimension
f = figure(figsize=(10,4))

# x position of Gaussian distribution
mu = np.mean(Z[class_mask,0]) 
mu2 = np.mean(Z[class_mask,1])

# Variance
variance = np.var(Z[class_mask,0])
variance2 = np.var(Z[class_mask,1])

# Spread-out-ness of bell shape
sigma = np.sqrt(variance)   
sigma2 = np.sqrt(variance2)

x = np.linspace(-10, 5, 100)

# Plot of variance and Gaussian distribution in same plot
plot(Z[class_mask,1], np.zeros_like(Z[class_mask,1]), 'bx') 
plot(Z[class_mask,0], np.zeros_like(Z[class_mask,0]), 'gx')
plot(x,mlab.normpdf(x, mu, sigma), 'g', x, mlab.normpdf(x, mu2, sigma2), 'b')
plt.title("Variance and distributions for PC1 and PC2")
plt.xlabel('Principal components')
show()
```

### Visualizing digits

```python
# Visualize the reconstructed data from the first K principal components
# Select randomly D digits.
figure(figsize=(10,5))
# Spread out-ness
W = np.dot(Z[:,range(K)],V[:,range(K)].T)
D = len(nD)
for d in range(D):
    digit_ix = np.random.randint(0,N)
    
    subplot(2, D, d+1)
    I = np.reshape(X[digit_ix,:], (16,16))
    imshow(I, cmap=cm.gray_r)
    title('Original')
    
    subplot(2, D, D+d+1)
    I = np.reshape(W[digit_ix,:]+X.mean(0), (16,16))
    imshow(I, cmap=cm.gray_r)
    title('Reconstr.'); 
```

#### What happens when analysing several digits?

They get harder to distinguish as many of their attributes will overlap.

```python
# Visualize the pricipal components
figure(figsize=(7,10))
for k in range(K):
    N1 = np.ceil(np.sqrt(K)); N2 = np.ceil(K/N1)
    subplot(N2, N1, k+1)
    I = np.reshape(V[:,k], (16,16))
    imshow(I, cmap=cm.hot)
    title('PC{0}'.format(k+1));

# output to screen
show()
```

#### How many principal components are needed to see the different digits properly?

With almost same color: 100

Otherwise 50 will do fine but in a more grainy grey-scale.
